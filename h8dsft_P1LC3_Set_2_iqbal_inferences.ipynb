```python
with open ('pca.pkl', 'rb') as file_1:
    pca2 = pickle.load(file_1)
with open ('kmeans.pkl', 'rb') as file_2:
    km = pickle.load(file_2)

with open ('winsoriser.pkl', 'rb') as file_3:
    winsorizer = pickle.load(file_3)
with open ('scaler.pkl','rb') as file_4:
    scaler = pickle.load(file_4)
```

# 10. Model Inferences

```python
data_inferences = pd.read_csv('humanitarian-aid.csv')
```

```python
data_inferences = data_inferences.head(10)
```

```python
df = data_inferences.copy()
```

```python
data_inferences
```

```python
data_inferences.drop('country' , inplace=True , axis=1)
```

```python
data_inferences
```

```python
winsorizer = Winsorizer(capping_method='iqr', tail='both', fold=1.5, variables=['child_mortality', 'exports', 'health', 'imports', 'income', 'inflation', 'life_expectancy', 'total_fertility', 'gdpp'])
winsorizer.fit(data_inferences)
data_inferences = winsorizer.transform(data_inferences)
```

```python
data_inferences=scaler.transform(data_inferences)
```

```python
data_inferences=pca2.transform(data_inferences)
```

```python
data_inferences
```

```python
reduce_data_inferences = pd.DataFrame(data_inferences, columns=['PC1', 'PC2', 'PC3' , 'PC4','PC5'])
reduce_data_inferences.head()
```

```python
inf_pred = km.predict(reduce_data_inferences)
```

```python
reduce_data_inferences['clusters'] = inf_pred
reduce_data_inferences.head()
```

```python
concat = reduce_data_inferences.head(10)
```

```python
final_inf = pd.concat([df, concat], axis=1)
```

```python
final_inf
```
